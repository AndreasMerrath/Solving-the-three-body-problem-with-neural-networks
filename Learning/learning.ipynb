# This File shows how to use the find_minimum file

```python
import find_minimum as fm


def initValues():
    tau = 0.01
    n = 3

    r = [[-7.673580434738827E-03, -3.302239590214931E-03, 2.073556671274906E-04],
         [3.452597940573748E-01, -1.498122010844959E-01, -4.413780254122306E-02],
         [1.031676785783527E-01, -7.220322696819025E-01, -1.605894448118791E-02]]

    v = [[5.284080958425666E-06, -6.612725729575814E-06, -5.777615059407366E-08],
         [5.305044905894823E-03, 2.724866176413712E-02, 1.741157849018483E-03],
         [1.986026764570113E-02, 3.004963198838360E-03, -1.104347674861390E-03]]

    m = [1, 3.302e23 / 1988500e24, 48.685e23 / 1988500e24]

    return tau, n, m, r, v

# import os
# 
# os.environ['TF_CPP_MIN_LOG_LEVEL'] = '1'
```

### Now, let's learn a few values

#### First, the mass of a body. This is expressed by giving "m" as a parameter

```python
tau, n, m, r, v = initValues()
values, costs, gradients = fm.find_global_minimum(
    tau, n, m, r, v,
    learning_rate=0.1,
    startingPosition=100,
    num_of_steps_in_do_step=10,
    accuracy=1e-5,
    variable_to_learn="m"
)
print("The learned values for each epoch are: ", values)
fm.plot(costs, "costs")
```

#### Then the position.This is expressed by giving "r" as a parameter

```python
tau, n, m, r, v = initValues()
values, costs, _ = fm.find_global_minimum(
    tau, n, m, r, v,
    learning_rate=0.1,
    startingPosition=[100, -50, 1],
    num_of_steps_in_do_step=25,
    accuracy=0.000000001,
    variable_to_learn="r"
)

print("The learned values for each epoch are: ", values)
fm.plot(costs, "costs")
```

#### Then the velocity. This is expressed by giving "v" as a parameter

```python
tau, n, m, r, v = initValues()

values, costs, _ = fm.find_global_minimum(
    tau, n, m, r, v,
    learning_rate=0.1,
    startingPosition=[100, -50, 0],
    num_of_steps_in_do_step=25,
    accuracy=0.000000001,
    variable_to_learn="v"
)
print("The learned values for each epoch are: ", values)
fm.plot(costs, "costs")
```

#### Now, all Masses at once.

```python
tau, n, m, r, v = initValues()
values, costs, _ = fm.learn_all_masses_Simulated_Values(
    tau, n, m, r, v,
    learning_rate=0.1,
    starting_guess=[1692, -342, 0.00000012301],
    num_of_steps_in_do_step=25,
    accuracy=0.00000001,
    convergence_rate=1.001,
    learning_rate_factor=8
)
print("The learned values for each epoch are: ", values)
fm.plot(costs, "costs")
```
